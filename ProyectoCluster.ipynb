```python
import pandas as pd
df = pd.read_excel('/content/TotalML.xlsx') # Use read_excel to read Excel files
df
```

```python
df.describe()
```

```python
# Filtrar los datos
df_0 = df[df['Promedio_Dias_Hasta_Siniestro'] == 0]
df_not_0 = df[df['Promedio_Dias_Hasta_Siniestro'] != 0]
```

```python
# Import the seaborn library
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración general de los gráficos
sns.set(style="whitegrid")

# Crear una figura con dos subgráficos
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico de distribución para cuando es 0
sns.histplot(df_0['Promedio_Dias_Hasta_Siniestro'], kde=True, ax=axes[0], color='blue')
axes[0].set_title('Distribución de Promedio_Dias_Hasta_Siniestro (0)')

# Gráfico de distribución para cuando es distinto de 0
sns.histplot(df_not_0['Promedio_Dias_Hasta_Siniestro'], kde=True, ax=axes[1], color='orange')
axes[1].set_title('Distribución de Promedio_Dias_Hasta_Siniestro (Distinto de 0)')

plt.tight_layout()
plt.show()
```

```python
# Crear un boxplot para `Tipo_Cartera`
plt.figure(figsize=(12, 6))
sns.boxplot(x='Tipo_Cartera', y='Promedio_Dias_Hasta_Siniestro', data=df_not_0)
plt.title('Distribución de Promedio_Dias_Hasta_Siniestro por Tipo_Cartera (Distinto de 0)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Crear un boxplot para `Producto`
plt.figure(figsize=(20, 9))
sns.boxplot(x='Producto', y='Promedio_Dias_Hasta_Siniestro', data=df_not_0)
plt.title('Distribución de Promedio_Dias_Hasta_Siniestro por Producto (Distinto de 0)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identificar variables cualitativas (categóricas) y cuantitativas
categorical_vars = df.select_dtypes(include=['object', 'category']).columns
quantitative_vars = df.select_dtypes(include=['number']).columns

# Graficar la distribución de variables cualitativas (categóricas)
plt.figure(figsize=(16, len(categorical_vars) * 5))  # Ajustar el tamaño de la figura según la cantidad de variables
for i, var in enumerate(categorical_vars, 1):
    plt.subplot(len(categorical_vars), 1, i)
    sns.countplot(data=df, x=var, palette="Set2")
    plt.title(f'Distribución de {var}')
    plt.xticks(rotation=45)
    plt.tight_layout()

# Mostrar las gráficas de variables cualitativas
plt.show()

# Graficar la distribución de variables cuantitativas
plt.figure(figsize=(16, len(quantitative_vars) * 5))  # Ajustar el tamaño de la figura según la cantidad de variables
for i, var in enumerate(quantitative_vars, 1):
    plt.subplot(len(quantitative_vars), 1, i)
    sns.histplot(df[var], kde=True, color="skyblue")
    plt.title(f'Distribución de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.tight_layout()

# Mostrar las gráficas de variables cuantitativas
plt.show()
```
